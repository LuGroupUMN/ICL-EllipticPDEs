```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

## Introduction:
In this workflow, we display how we generate the scaling law plot with respect to parameter $N$ in Fig 1. A.

```python
def generate_colors(n, colormap='coolwarm'): #'viridis'
    cmap = plt.get_cmap(colormap)
    colors = [cmap(i / n) for i in range(n)]
    return colors
```

```python
N_test_np = np.load("N_test.npy")
```

```python
columns_1 = ["d", "n", "N", "N_test", "m", "err", "err_log"]
N_test_pd = pd.DataFrame(data=N_test_np, columns=columns_1)
N_test_pd["m"] = N_test_pd["m"].apply(lambda x: int(x))
```

```python
fig, ax = plt.subplots(figsize=(10,6))
m_list = [5000 ,10000,20000] 
markers_list = ['o', '^', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_']

x_np_rate = np.array([i for i in range(0,10000,100)])
color_list = generate_colors(len(m_list))
intercept_list = []
slop_list = []
shift_list = []
for m_index in range(len(m_list)):
    c_select_1 = N_test_pd["n"]==500
    c_select_2 = N_test_pd["d"]==5
    c_select_3 = N_test_pd["m"]==m_list[m_index]
    c_select_4 = N_test_pd["N"]<=600
    c_select = c_select_1 & c_select_2 & c_select_3 & c_select_4
    
    data = N_test_pd[c_select]
    data_np_y = list(np.array(data["err"]))
    data_np_x = list(np.array(data["N"]))
    
    slope_temp, intercept_temp = np.polyfit(np.log10(data_np_x), np.log10(data_np_y), 1)

    y_np = np.log10(np.array(data_np_x))* (-0.5)
    shift_term = np.mean(np.log10(data_np_y)) - np.mean(y_np)

    shift_list.append(shift_term)
    intercept_list.append(intercept_temp)
    slop_list.append(slope_temp)
    ax.scatter(data_np_x, data_np_y, marker=markers_list[m_index], color=color_list[m_index], label = "m="+str(m_list[m_index]))
    
y_np =  10 ** (shift_list[1] + np.log10(np.array(data_np_x))* (-0.5))

ax.plot(data_np_x, y_np, "--", color="r", label=r"O(N$^{-1/2}$)")

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('N', fontsize=20)
ax.set_ylabel('Error', fontsize=20)
ax.set_title('n=500, d=5', fontsize=20)
plt.legend(fontsize=10, loc =1)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

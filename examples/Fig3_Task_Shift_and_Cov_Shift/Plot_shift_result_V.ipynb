```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_shift_V = pd.read_csv("testdistribution_V.csv",index_col=False)
left_right_pairs = [[1,2], [2,3], [3,5], [5,10], [10,20]]
```

```python
data_shift_V
```

```python
f, ax = plt.subplots(figsize=(6,6))
sns.lineplot(data=data_shift_V[data_shift_V["n"]==300], x="m",y="err_h1_relative",err_style="bars",markers=True, hue="index",style="index",linewidth=2, palette="coolwarm") #sns.color_palette("husl"))

ax.set(xscale="log", yscale="log")
legend = plt.legend(fontsize=12, loc = 1)
i = 0
for text in legend.texts:    
    text.set_text(r"$V(x)\sim \mathbf{U}(" + str(left_right_pairs[i][0]) + ", " + str(str(left_right_pairs[i][1])) + ")$")
    i += 1

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Context Length (m)", fontsize=20)
plt.ylabel("Relative H1 Error", fontsize=20)
plt.title("d=50, n=300", fontsize=20)
```

```python
f, ax = plt.subplots(figsize=(6,6))
sns.lineplot(data=data_shift_V[data_shift_V["n"]==300], x="m",y="err_l2_relative",err_style="bars",markers=True, hue="index",style="index",linewidth=2, palette="coolwarm") #sns.color_palette("husl"))

ax.set(xscale="log", yscale="log")

legend = plt.legend(fontsize=12, loc = 1)
i = 0
for text in legend.texts:    
    text.set_text(r"$V(x)\sim \mathbf{U}(" + str(left_right_pairs[i][0]) + ", " + str(str(left_right_pairs[i][1])) + ")$")
    i += 1
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Context Length (m)", fontsize=20)
plt.ylabel("Relative L2 Error", fontsize=20)
plt.title("d=50, n=300", fontsize=20)
```
